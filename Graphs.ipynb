```python
import pandas as pd 
import numpy as np
```

```python
data1 = pd.read_csv("Fuzzing/output.csv",header=None)
data2 = pd.read_csv("BasicLearning/output.csv",header=None)
data3 = pd.read_csv("CombinedLearningFuzzing/output.csv",header=None)

reachProblems = ["Problem11", "Problem12", "Problem13", "Problem14", "Problem15", "Problem16","Problem17","Problem18", "Problem19"]
```

```python
errors_fuzzing = data1[data1[0].isin(reachProblems)]
errors_basiclearning = data2[data2[0].isin(reachProblems) & (data2[1] == "learning_error")]
errors_combinedlearning = data3[data3[0].isin(reachProblems) & (data3[1] == "learning_error")]
```

```python
errors_basiclearning[1] = "Basic Learning"
errors_combinedlearning[1] = "Combined Learning"
errors_fuzzing[1] = "Fuzzing"
```

```python
result_df = pd.concat([errors_fuzzing, errors_basiclearning, errors_combinedlearning])
```
